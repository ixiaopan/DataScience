```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 50, 0.1)
n = len(x)
    
a = np.random.randn(n, 1)
b = np.random.randn(n, 1)
c = np.random.randn(n, 1)

def getKCov(l=1):
    K = np.zeros((n, n))

    for i in range(n):
        K[i, i] = 1
        for j in range(0, i):
            K[j, i] = K[i, j] = np.exp(-(x[i] - x[j])**2/(2*l))

    return K




def plot(K):
    L = np.linalg.cholesky(K + np.eye(n) * 0.001)

    f1 = L @ a
    f2 = L @ b
    f3 = L @ c

    fig, ax = plt.subplots(figsize=(8, 2))
    
    ax.plot(x, f1)
    ax.plot(x, f2)
    ax.plot(x, f3)
    ax.set_ylim(-4, 4)

    plt.show()
```

```python

plot(getKCov(1))
```

```python
plot(getKCov(4))
```
